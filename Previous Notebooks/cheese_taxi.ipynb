```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
```

```python
trainset = pd.read_csv("processed_taxi/train.csv")
testset = pd.read_csv("processed_taxi/test.csv")
```

```python
print(trainset.columns)
```

```python
print(testset.columns)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

from torch.utils.data import TensorDataset, DataLoader
```

```python
from sklearn.model_selection import train_test_split
validset, validset = train_test_split(trainset, test_size=0.2, random_state=42)
```

```python
device = torch.device("mps")
```

```python
split_col = "TAXI_ID"
split_col_len = 981
```

```python
X_trains = [torch.tensor(trainset.drop(columns=["TRIP_ID", "LEN", "POLYLINE", "TIMESTAMP"])[trainset[split_col] == i].values, dtype=torch.float32).to(device) for i in range(split_col_len)]
y_trains = [torch.tensor(trainset[trainset[split_col] == i]["LEN"].values, dtype=torch.float32).to(device) for i in range(split_col_len)]
```

```python
X_valids = [torch.tensor(validset.drop(columns=["TRIP_ID", "LEN", "POLYLINE", "TIMESTAMP"])[validset[split_col] == i].values, dtype=torch.float32).to(device) for i in range(split_col_len)]
y_valids = [torch.tensor(validset[trainset[split_col] == i]["LEN"].values, dtype=torch.float32).to(device) for i in range(split_col_len)]
```

```python
for i in range(split_col_len):
    print(X_trains[i].shape, y_trains[i].shape, X_valids[i].shape, y_valids[i].shape)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

general = LinearRegression().fit(trainset.drop(columns=["TRIP_ID", "LEN", "POLYLINE", "TIMESTAMP"]).values, trainset["LEN"].values)

models = [general if X_trains[i].shape[0] == 0 else LinearRegression().fit(X_trains[i].cpu().numpy(), y_trains[i].cpu().numpy()) for i in range(split_col_len)]
```

```python
with torch.no_grad():
    
    valid_losses = [0 if X_valids[i].shape[0] == 0 else mean_squared_error(torch.max(torch.tensor(models[i].predict(X_valids[i].cpu().numpy())), torch.zeros(*y_valids[i].shape)).numpy(), y_valids[i].cpu().numpy())*X_valids[i].shape[0] for i in range(split_col_len)]
    mse = sum(valid_losses)/sum([X_valids[i].shape[0] for i in range(split_col_len)])
    print(mse**0.5)
```

```python
x_test = validset.drop(columns=["TRIP_ID", "POLYLINE", "LEN", "TIMESTAMP"])
outputs = [max(0, models[x_test[split_col][i]].predict(torch.tensor(row).reshape(1, -1))[0]) for i, row in x_test.iterrows()]
t = np.array([])
final_predictions = pd.DataFrame().assign(
    TRIP_ID= validset["TRIP_ID"],
    TRAVEL_TIME = outputs
    )

final_predictions
```

```python
print(final_predictions["TRAVEL_TIME"].value_counts()/len(final_predictions["TRAVEL_TIME"]))
print(final_predictions["TRAVEL_TIME"].mean())
```

```python
x_test = testset.drop(columns=["TRIP_ID", "TIMESTAMP"])
offset = 883.6303416117978 - 716.6585632336948
outputs = [max(0, models[x_test[split_col][i]].predict(torch.tensor(row).reshape(1, -1))[0] - offset) for i, row in x_test.iterrows()]
t = np.array([])
final_predictions = pd.DataFrame().assign(
    TRIP_ID= testset["TRIP_ID"],
    TRAVEL_TIME = outputs
    )

final_predictions
```

```python
print(final_predictions["TRAVEL_TIME"].value_counts()/len(final_predictions["TRAVEL_TIME"]))
```

```python
print(final_predictions["TRAVEL_TIME"].mean())
```
